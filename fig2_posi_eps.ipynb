## Figure2. Level curves of the first integrals for $\epsilon>0$
- parameters : $\alpha=0.1, \beta = 0.01, \delta=0.02, \gamma=0.1, \epsilon=0.005, \lambda=0.01$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
from PIL import Image
import math
```

```python
plt.rcParams["figure.figsize"] = (6,4.5)
```

```python
np.seterr(divide='ignore', invalid='ignore') # avoiding invalid value encountered in true_divide
```

```python
# Define the system of ODEs
# x is prey, y is predator

def first_integral(x, y):
    # base params
    alpha = 0.1; beta = 0.01; 
    delta = 0.02; gam = 0.1;
    eps = 0.005; lamda = 0.01;
    
#     # original model
#     alpha = 0.1; beta = 0.01; 
#     delta = 0.02; gam = 0.1;
#     eps = 0; lamda = 0;
    
    return (alpha*np.log(y) - beta*y + eps*delta*x*y - eps*gam*y - (alpha*lamda +delta)*x + gam*np.log(x))
```

```python
# check the condition
alpha = 0.1; beta = 0.01; 
delta = 0.02; gam = 0.1;
eps = 0.005; lamda = 0.01;

# # original model
# alpha = 0.1; beta = 0.01; 
# delta = 0.02; gam = 0.1;
# eps = 0; lamda = 0;

print(beta*lamda - eps*delta)
```

```python
# p = np.linspace(0, 20, 100)
# q = np.linspace(0, 20, 100)

p = np.linspace(0, 30, 100)
q = np.linspace(0, 30, 100)

X1 , Y1 = np.meshgrid(p,q)
```

```python
z = first_integral(X1, Y1)
```

```python
contours = plt.contour(X1, Y1, z, levels = 6, cmap='rainbow')
plt.xlabel("Prey", fontsize = 15)
plt.ylabel("Predator", fontsize = 15)
plt.savefig('re_image/fig2-2.eps', format='eps')
plt.show()
```
